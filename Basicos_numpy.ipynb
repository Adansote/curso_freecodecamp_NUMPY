**Inicialización de diferentes tipos de matrices**

**# Matriz de todos ceros**

```python
np.zeros((2,3))
```

```python
## Matriz de todos 1
np.ones((4,2,2), dtype='int32')
```

```python
#cualquier otro numero
np.full_like((2,2),99)
```

```python
#cualquier otro numero(full_like)
np.full_like(a,4)
```

```python
# Números decimales aleatorios
np.random.rand(4,2)
```

```python
# Valores enteros aleatorios
np.random.randint(-4,8, size=(3,3))
```

```python
# La matriz de identidad
np.identity(5)
```

```python
# Repetir una matriz
arr = np.array([[1,2,3]])
r1 = np.repeat(arr,3, axis=0)
print(r1)
```

```python
output = np.ones((5,5))
print(output)

```

```python
z = np.zeros((3,3))
z[1,1] = 9
print(z)
```

```python
output[1:-1,1:-1] = z
print(output)
```

**Reorganización de matrices**

```python
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)
```

```python
after = before.reshape((2,3))
print(after)
```

```python
# Vectores apilados verticalmente
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.vstack([v1, v2, v1, v2])
```

```python
# Pila horizontal
h1 = np.ones((2,4))
h2 = np.zeros((2,2))

np.hstack((h1,h2))
```
